```python
import pandas as pd
import plotly.express as px
```

```python
energy_df = pd.read_csv("../data/energy/energy_dataset.csv")
energy_df.head()
```

```python
energy_df["time"] = pd.to_datetime(energy_df.time, utc=True)
```

```python
energy_df.head()
```

```python
energy_df.dtypes
```

```python
px.line(data_frame=energy_df, x="time", y="generation fossil gas")
```

```python
px.line(data_frame=energy_df, x="time", y="total load actual")
```

```python
month_df = energy_df.groupby(energy_df.time.dt.month).sum()
month_df.head()
```

```python
px.line(data_frame=month_df, x=month_df.index, y="total load actual")
```

```python
px.pie(values=energy_df.describe().iloc[1], names=energy_df.describe().iloc[1].index)
```

```python
mean_generation_values = energy_df.describe().iloc[1].sort_values(ascending=False)
mean_generation_values = mean_generation_values[mean_generation_values>0]
mean_generation_values = mean_generation_values[:-2]
other = pd.Series(mean_generation_values[7:].sum(), index=["other"])
mean_generation_values = mean_generation_values[:7].append(other)
mean_generation_values
```

```python
px.pie(values=mean_generation_values, names=mean_generation_values.index)
```

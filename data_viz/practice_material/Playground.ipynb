```python
# mass_wrt_earth
# density = mass / kg
# volume_wrt_earth
```

```python
exoplanet_df["conv_factor_r"] = exoplanet_df["radius_wrt"].map({"Jupiter": 11, "Earth": 1})
```

```python
exoplanet_df["raidus_wrt_earth"] = exoplanet_df["conv_factor_r"]*exoplanet_df["radius_multiplier"]
```

```python
exoplanet_df["volume"] = 4/3 * 3.14 * exoplanet_df["raidus_wrt_earth"]**3
```

```python
exoplanet_df["density"] = exoplanet_df["total_mass_wrt_earth"]/exoplanet_df["volume"]
```

## Methods & libraries

Python is so powerful because of its open-source community that everyday develops new "methods" for people like you and me to use and save hours of our time. Methods simply are bits of code that have been written by someone else and published online for others to install and reuse. There are methods for (almost) EVERYTHING! 

```python
# A method is applied on a variable using the following syntax
name_of_variable.method()
```

```python
# Python comes with many "built-in" methods like this one below
first_name = "Marie"
first_name.upper()
```

```python
# Or that one
first_name.capitalize()
```

```python
# Some methods require parameters, like this one
last_name = "Guttenberg"
last_name.count("e")
```

A library is just a collection of methods. Usually libraries would be task-specific. There are libraries specifically developed from data visualization (that we'll use in a moment) and others for machine learning. Some libraries are very niche like the `lbl` library that was developed for computing radial velocities of red dwarfs.

# Data visualization

## Libraries to import

```python
# Once installed, libraries need to be imported inside your workspace
import pandas as pd
import numpy as np
import plotly.express as px
```

## Define path to data

```python
file_path = "data/CA.csv"
```

```python
dataframe = pd.read_csv(file_path)
dataframe.columns = dataframe.iloc[0]
dataframe.drop(index=0, inplace = True)
```

```python
dataframe
```

```python
dataframe.dtypes
```

```python
dataframe["DISTANCE FROM SHORE"] = dataframe["DISTANCE FROM SHORE"].astype(float)
```

```python
dataframe["MEAN DEPTH"] = dataframe["MEAN DEPTH"].astype(float)
```

```python
fig = px.scatter(x=dataframe["DISTANCE FROM SHORE"], y=dataframe["MEAN DEPTH"])
fig.update_yaxes(autorange="reversed")
```
